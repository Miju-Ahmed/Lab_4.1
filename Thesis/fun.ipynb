```python
import time

def rocket_launch_simulation():
    print("🚀 Initializing pre-launch checks...")
    time.sleep(1)
    
    # Fictional safety checks
    for check in ["Fuel OK", "Navigation OK", "Payload Secure", "Weather Clear"]:
        print(f"🔍 {check}")
        time.sleep(0.5)
    
    # Countdown
    print("\nCountdown started!")
    for i in range(10, 0, -1):
        print(f"{i}...")
        time.sleep(1)
    
    # Launch!
    print("\nIGNITION! LIFTOFF! 🚀🔥")
    print('''
        /\\
       /  \\
      /____\\
      |####|
      |####|
    ''')

    # Simulate ascent
    for altitude in range(0, 10000, 1000):
        print(f"Altitude: {altitude} km...")
        time.sleep(0.3)

    print("\nOrbit achieved! 🌍🛰️")

if __name__ == "__main__":
    user_input = input("Type 'LAUNCH' to start: ").strip().upper()
    if user_input == "LAUNCH":
        rocket_launch_simulation()
    else:
        print("Launch aborted. Safety first! 🔒")
```

```python
import time
import os

def countdown(n):
    for i in range(n, 0, -1):
        print(f"\rT-{i} seconds", end='', flush=True)
        time.sleep(1)
    print("\n🚀 Liftoff! 🚀")

def launch_rocket():
    rocket = [
        "    ▲    ",
        "   ▲▲▲   ",
        "  ▲▲▲▲▲  ",
        " ▲▲▲▲▲▲▲ ",
        "    █    ",
        "    █    ",
        "    █    "
    ]
    sky = [" "] * 20

    for i in range(len(sky) + len(rocket)):
        os.system("clear" if os.name == "posix" else "cls")  # Clear screen
        for j in range(max(0, i - len(rocket)), i):
            print(sky[j])
        for line in rocket:
            print(line)
        time.sleep(0.2)

    print("🚀 The rocket is in space! 🌌")

# Start launch sequence
countdown(5)
launch_rocket()
```
