```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Erosion and Dialation Kernel

```python
image = np.zeros((100, 100), dtype=np.uint8)
cv2.rectangle(image, (20, 20), (80, 80), 255, -1)

kernel = np.ones((5, 5), np.uint8) 

eroded_image = cv2.erode(image, kernel, iterations=1)

dilated_image = cv2.dilate(image, kernel, iterations=1)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Eroded Image")
plt.imshow(eroded_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Dilated Image")
plt.imshow(dilated_image, cmap='gray')

plt.tight_layout()
plt.show()
```

Make Erosion and Dialation

Eroted Image

```python
image = cv2.imread('Images/m1.jpeg', cv2.IMREAD_GRAYSCALE)
image=cv2.resize(image,(512,512))

_, binary_image = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
plt.title("Original Binary Image")
plt.imshow(binary_image, cmap='gray')

kernel = np.ones((3, 3), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 2)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((5, 5), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 3)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((9, 9), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 4)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((11, 11), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 5)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((15, 15), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 6)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

plt.tight_layout()
plt.show()
```

Dialation

```python
image = cv2.imread('Images/m1.jpeg', cv2.IMREAD_GRAYSCALE)
image=cv2.resize(image,(512,512))

_, binary_image = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
plt.title("Original Binary Image")
plt.imshow(binary_image, cmap='gray')

kernel = np.ones((3, 3), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 2)
plt.title("Dialated Image")
plt.imshow(dilated_image, cmap='gray')

kernel = np.ones((5, 5), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 3)
plt.title("Dialated Image")
plt.imshow(dilated_image, cmap='gray')

kernel = np.ones((9, 9), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 4)
plt.title("Dialated Image")
plt.imshow(dilated_image, cmap='gray')

kernel = np.ones((11, 11), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 5)
plt.title("Dialated Image")
plt.imshow(dilated_image, cmap='gray')

kernel = np.ones((15, 15), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 6)
plt.title("Dialated Image")
plt.imshow(dilated_image, cmap='gray')

plt.tight_layout()
plt.show()
```
