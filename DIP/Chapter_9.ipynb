Erosion and Dialation Kernel

```python
image = np.zeros((100, 100), dtype=np.uint8)
cv2.rectangle(image, (20, 20), (80, 80), 255, -1)

kernel = np.ones((5, 5), np.uint8) 

eroded_image = cv2.erode(image, kernel, iterations=1)

dilated_image = cv2.dilate(image, kernel, iterations=1)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Eroded Image")
plt.imshow(eroded_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Dilated Image")
plt.imshow(dilated_image, cmap='gray')

plt.tight_layout()
plt.show()
```

Eroted Image

```python
image = cv2.imread('Images/m1.jpeg', cv2.IMREAD_GRAYSCALE)
image=cv2.resize(image,(512,512))

_, binary_image = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
plt.title("Original Binary Image")
plt.imshow(binary_image, cmap='gray')

kernel = np.ones((3, 3), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 2)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((5, 5), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 3)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((9, 9), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 4)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((11, 11), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 5)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

kernel = np.ones((15, 15), np.uint8)
eroded_image1 = cv2.erode(binary_image, kernel, iterations=1)
plt.subplot(2, 3, 6)
plt.title("Eroded Image")
plt.imshow(eroded_image1, cmap='gray')

plt.tight_layout()
plt.show()
```

Closing

```python
image = cv2.imread('Images/open.png', cv2.IMREAD_GRAYSCALE)

_,binary_image = cv2.threshold(image,127, 255, cv2.THRESH_BINARY)
kernel = np.ones((49,49), np.uint8)

eroded_image = cv2.dilate(binary_image, kernel, iterations=1)
eroded_image = cv2.erode(eroded_image, kernel, iterations=1)

closed_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.title("Original Binary Image")
plt.imshow(binary_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Closed Image")
plt.imshow(closed_image, cmap='gray')

plt.tight_layout()
plt.show()
```

Edge detect or Boundary Extraction Algorithm
b(A) = A - (A erosion B)

```python
image = cv2.imread('Images/open.png', cv2.IMREAD_GRAYSCALE)
_,binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

kernel = np.ones((3,3), np.uint8)

er = cv2.erode(binary_image, kernel, iterations=1)
er = binary_image - er

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.title("Original Binary Image")
plt.imshow(binary_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Boundary Extraction")
plt.imshow(er, cmap='gray')

plt.tight_layout()
plt.show()
```

Convex Hull

```python
image = cv2.imread('Images/M1.jpeg', cv2.IMREAD_GRAYSCALE)

_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

hull_image = cv2.cvtColor(binary_image, cv2.COLOR_GRAY2BGR)

# for contour in contours:
#     hull = cv2.convexHull(contour)  # Compute the convex hull
#     cv2.drawContours(hull_image, [hull], -1, (0, 255, 0), 2)  # Draw the convex hull

for contour in contours:
    if cv2.contourArea(contour) > 10:  # Filter small contours
        hull = cv2.convexHull(contour)
        cv2.drawContours(hull_image, [hull], -1, (0, 255, 0), 2)

# Display the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.title("Original Binary Image")
plt.imshow(binary_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Convex Hull")
plt.imshow(hull_image)

plt.tight_layout()
plt.show()
```
