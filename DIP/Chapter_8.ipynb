```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
from scipy.signal import convolve2d, correlate2d
```

Histogram

```python
image=cv2.imread("Images/ps.jpeg", cv2.IMREAD_GRAYSCALE)
th_image=cv2.imread("Images/ps1.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12,6))
plt.subplot(2,2,1),plt.title("Original Image"),plt.imshow(image,cmap='gray')
plt.subplot(2,2,2),plt.title("Another Image"),plt.imshow(th_image,cmap='gray')
plt.subplot(2,2,3),plt.title("Original Image"),plt.bar(range(256), cv2.calcHist([image], [0], None, [256], [0, 256]).ravel())
plt.subplot(2,2,4),plt.title("Another Image"),plt.bar(range(256),cv2.calcHist([th_image],[0],None,[256],[0,256]).ravel())
plt.tight_layout()
plt.show()
```
